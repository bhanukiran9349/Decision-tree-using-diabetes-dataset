```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data=pd.read_csv("diabetes.csv")
```

```python
data
```

```python
data.isna()
```

```python
data.isna().sum()
```

```python
data.columns
```

```python
#feature selection
#columns are caller attributes or features
x=pd.DataFrame(data,columns=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'])
y=pd.DataFrame(data,columns=['Outcome'])
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
#creating object
clf=DecisionTreeClassifier()

#train the decision tree classiifier
clf=clf.fit(x_train,y_train)
#predict the response
y_pred=clf.predict(x_test)
```

```python
from sklearn import tree
tree.plot_tree(clf)


import matplotlib.pyplot as plt 
fn=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
cn=['0', '1']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(clf,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('imagename.png')
```

```python
print('Accuracy:',metrics.accuracy_score(y_test,y_pred))
```
